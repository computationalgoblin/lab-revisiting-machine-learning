```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\angel\Desktop\Data Analysis\Ironhack\Labs\lab-revisiting-machine-learning\data_7.01_activities\files_for_activities\learningSet.csv\learningSet.csv")
data.head()
```

```python
# Extracting the categorical columns

categorical = data.select_dtypes("object")
categorical.head()
```

```python
# Checking nulls in categorical data
nulls = categorical.isnull().sum()
nulls[nulls.values != 0]
```

```python
# Vars to exclude 
vars_to_exclude = ["OSOURCE", "ZIP CODE"]
```

```python
# Adding the variables to a drop list
drop_list = []

for col in categorical.columns:
    if col not in vars_to_exclude:
        drop_list.append(col)

categorical = categorical[drop_list]
categorical.head()
print(categorical.columns)
```

```python
# Identify the columns that are over 85% of missing values

nulls_percents = nulls.values * 100 / categorical.shape[0]
nulls_percents
```

```python
# Reducing the number of categories in gender column
categorical["GENDER"].value_counts()
categorical["GENDER"] = categorical["GENDER"].fillna("F")
```

```python
categorical["GENDER"] = categorical["GENDER"].apply(lambda x: "other" if x not in ["M", "F"] else x)
categorical["GENDER"].value_counts()
```
